```python
from papersys.ocr import ocr_by_id_batch, response2md
from papersys.const import DATA_DIR, BASE_DIR
from dotenv import load_dotenv
import json

load_dotenv(BASE_DIR / ".env")
```

```python
# responses = ocr_by_id_batch(["2510.20810", "2510.20766"])
# with open(DATA_DIR / "ocr_responses_example.json", "w") as f:
#     json.dump(responses, f)
with open(DATA_DIR / "ocr_responses_example.json", "r") as f:
    responses = json.load(f)
```

```python
response_dir = DATA_DIR / "ocr_responses_example"
for arxiv_id, response in responses.items():
    cur_dir = response_dir / arxiv_id
    file_name = f"{arxiv_id}.md"
    response2md(response, cur_dir, file_name)
```
