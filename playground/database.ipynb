```python
manager2 = PaperManager(DATA_DIR / "papersys") 
```

```python
manager2.unembeded_papers2(["cs", "stat.ML"], use_scalar_index=False)
```

```python
manager2.unembeded_papers(["cs.CL", "stat.ML"])
```

```python
# manager2.db["paper_metadata"].search().select(["id", "categories"]).to_polars()

embedded_ids = manager2.embedding_table.search().select([ID]).to_polars()
all_ids = manager2.metadata_table.search().select([ID, CATEGORIES]).to_polars()
```

```python
embedded_ids.schema, all_ids.schema
```

```python
categories = ["cs", "stat.ML"]
```

```python
import polars as pl

# 筛选不在 embedded_ids 中的论文
not_embedded = all_ids.join(embedded_ids, on=ID, how="anti")

# 使用 Polars 原生表达式进行前缀匹配
# 对于每个类别，检查是否有任何 category 以它开头
filtered_ids = not_embedded.filter(
    pl.col(CATEGORIES).list.eval(
        pl.any_horizontal([
            pl.element().str.starts_with(cat) 
            for cat in categories
        ])
    ).list.any()
)

print(f"总共有 {len(all_ids)} 篇论文")
print(f"已嵌入 {len(embedded_ids)} 篇论文")
print(f"未嵌入且类别匹配的论文: {len(filtered_ids)} 篇")
filtered_ids
```

```python
a = filtered_ids.select(ID).to_series().to_arrow()
```

```python
# convert a to pa.string type
import pyarrow as pa
id_array = pa.array(a, type=pa.string())
id_array
```

```python
from tqdm import tqdm

def build_query_list(ids):
    core ="('" + "', '".join(ids) + "')"
    return core

batch_size = 128
id_list = filtered_ids.select(ID).to_series().to_list()

results = []    
for i in tqdm(range(0, len(id_list), batch_size)):
    batch_ids = id_list[i:i+batch_size]
    query_core = build_query_list(batch_ids)
    manager2.metadata_table.search().where(f"id IN {query_core}").to_polars()
    
```

```python
import pyarrow as pa
import pyarrow.compute as pc

ds = manager2.metadata_table.to_lance()
id_array = pa.array(id_list, type=pa.string())  # ids 为大量 id 列表
expr = pc.is_in(pc.field("id"), id_array)
pa_table = ds.to_table(filter=expr, use_scalar_index=False)
```

```python
import polars 
po = polars.from_arrow(pa_table)
```
