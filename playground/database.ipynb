```python
import papersys
from papersys.database.manager import PaperManager
from papersys.database.name import *
from papersys.database.schema import PAPER_METADATA_SCHEMA
from papersys.database.migrate import load_arxiv_oai_snapshot, fetch_arxiv_oai
from papersys.const import DATA_DIR
from datetime import date
```

```python
df = load_arxiv_oai_snapshot(
    DATA_DIR / "arxiv-metadata-oai-snapshot.jsonl",
)
df.head()
```

```python
fetch_results = fetch_arxiv_oai(
    categories=["cs"],
    start = date(2025,8,15),
    end = date(2025,9,1)
)

fetch_results.head()
```

```python
manager = PaperManager(DATA_DIR / "sample_database") 
manager.drop_metadata_table()
manager.create_metadata_table()
manager.db["paper_metadata"]
```

```python
from papersys.database.manager import upsert, add

t = manager.db["paper_metadata"]
add(t, df).head()
```

```python
manager.db["paper_metadata"].optimize()
```

```python
upsert(t, fetch_results).head()
```

```python
manager.db["paper_metadata"].head()
```

```python
table = manager.db["paper_metadata"]
results = table.search().where(
    """--sql
    (update_date BETWEEN to_date('2019-08-01') AND to_date('2019-12-31')) 
    AND array_has_any(categories, ['cs.LG']) 
    AND doi IS NOT NULL
    """
).to_polars()
results
```
