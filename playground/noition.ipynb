```python
import os
import ultimate_notion as uno
from dotenv import load_dotenv
from papersys.const import BASE_DIR

load_dotenv(BASE_DIR / ".env")

if not os.getenv("NOTION_TOKEN"):
    raise RuntimeError("NOTION_TOKEN is missing. Add it to your .env before continuing.")
```

```python
import uuid

def normalize_page_id(url_or_id: str) -> str:
    """Convert a Notion share link or bare id into API-friendly UUID."""
    trimmed = url_or_id.rsplit("/", 1)[-1]
    trimmed = trimmed.split("?")[0].split("#")[0]
    last_segment = trimmed.split("-")[-1]
    return str(uuid.UUID(last_segment))

page_url = "https://www.notion.so/Draft-27e81ffdcde380309eb3d9010ce98e4a"
page_id = normalize_page_id(page_url)

with uno.Session() as notion:
    page = notion.get_page(page_id)
    page.show()
```
